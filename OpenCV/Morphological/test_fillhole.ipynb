```python
import cv2 
import numpy as np

import matplotlib.pyplot as plt
```

```python
img = cv2.imread("kongdong.png")
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,img_bin = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
```

```python
hole_pos = (72,82)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
Xprev = np.zeros_like(img_bin)
Xprev[hole_pos[1],hole_pos[0]]=255
Xcurrent = cv2.bitwise_and(cv2.dilate(Xprev,kernel),np.array(255-img_bin,dtype='uint8'))
while(not (Xprev==Xcurrent).all()):
    Xprev = Xcurrent
    Xcurrent = cv2.bitwise_and(cv2.dilate(Xprev,kernel),np.array(255-img_bin,dtype='uint8'))
    plt.figure(1)
    plt.subplot(121)
    plt.imshow(Xprev,cmap='gray')
    plt.subplot(122)
    plt.imshow(255-img_bin,cmap='gray')
    plt.show()
```

```python
img_bin[Xcurrent==255] = 255
```

```python
plt.figure(figsize=(16,16))
cv2.circle(img,hole_pos,1,(0,255,0),-1)
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_bin,cmap='gray')
```
