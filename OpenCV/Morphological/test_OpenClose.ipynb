```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img1.png')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,img_bin = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
```

```python
## 开运算
kernel = np.ones((7,7),np.uint8)
img_open1 = cv2.morphologyEx(img_bin, cv2.MORPH_OPEN, kernel)
open_tmp = cv2.erode(img_bin,kernel)
img_open2 = cv2.dilate(open_tmp,kernel)

plt.figure(figsize=(16,16))
plt.subplot(131)
plt.imshow(img_bin,cmap='gray')
plt.subplot(132)
plt.imshow(img_open1,cmap='gray')
plt.subplot(133)
plt.imshow(img_open2,cmap='gray')
```

```python
## 闭运算
kernel = np.ones((7,7),np.uint8)
img_close1 = cv2.morphologyEx(img_bin, cv2.MORPH_CLOSE, kernel)
close_tmp = cv2.dilate(img_bin,kernel)
img_close2 = cv2.erode(close_tmp,kernel)

plt.figure(figsize=(16,16))
plt.subplot(131)
plt.imshow(img_bin,cmap='gray')
plt.subplot(132)
plt.imshow(img_close1,cmap='gray')
plt.subplot(133)
plt.imshow(img_close2,cmap='gray')
```
