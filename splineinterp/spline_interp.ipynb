```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
#创建数据点
x = np.arange(0,2*np.pi,1)
y = np.sin(x)
#被插值点
xx = np.arange(0,6,0.2)
```

```python
#python中的spline interpolation分为两步
#先计算spline表达式
tck=interpolate.splrep(x,y,s=0) #s是平滑程度
#再进行插值
yy=interpolate.splev(xx,tck,der=0)#der是导数
```

```python
plt.figure()
plt.plot(x, y, 'x', xx, yy, xx, np.sin(xx), x, y, 'b')
plt.legend(['linear','Spline','sin'])
plt.show()
```
