```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_point(x1,y1,x2,y2,x):
    return (x-x2)/(x1-x2)*y1+(x-x1)/(x2-x1)*y2

def linear_interp(x,y,xx):
    assert len(x)==len(y),'x dimension not match with y'
    yy = np.zeros((len(xx),),dtype=np.float32)
    for i in range(len(xx)):
        for j in range(len(x)):
            if(xx[i]>x[j] and xx[i]<=x[j+1]):
                yy[i] = calc_point(x[j],y[j],x[j+1],y[j+1],xx[i])
                continue
    return yy
```

```python
# 制作数据
x = np.linspace(0,2*np.pi,7) #数据集x坐标
y = np.sin(x) #数据集y坐标
xx = np.arange(0,6,0.5)#待预测数据集
## 预测
yy = linear_interp(x,y,xx)
```

```python
plt.plot(x,y,color="r", linestyle=":", marker="o", linewidth=1)
plt.scatter(xx,yy,color="g", marker="^")
```

```python
#使用numpy自带分段线性插值函数验证
yy_np = np.interp(xx,x,y)
```

```python
plt.plot(x,y,color="r", linestyle=":", marker="o", linewidth=1)
plt.scatter(xx,yy,color="g", marker="^")
plt.scatter(xx,yy_np,color="b", marker="*")
```
